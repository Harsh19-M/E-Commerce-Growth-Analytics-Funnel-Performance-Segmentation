```python
# TEST 1 
'''
H1: Does Variant B have a higher conversion rate than Variant A?
1️⃣ Hypothesis (what we’re testing)
H0 (Null): Conversion rate of B ≤ Conversion rate of A (no improvement).
H1 (Alternative): Conversion rate of B > Conversion rate of A (B performs better).

2️⃣ How we’ll do it (steps)
Take the analysisECOM_lvl table.

Group by experiment_variant (A or B) and calculate:
Number of users shown the variant (n_users)
Number of users who converted (n_converted)
Conversion rate = n_converted / n_users
Use a proportion z-test (from statsmodels) to see if the difference in conversion rate is statistically significant.
Pick a significance level (α = 0.05, standard in companies).

3️⃣ How we interpret the result
p-value < 0.05 → Reject H0 → Variant B is significantly better than A 
p-value ≥ 0.05 → Fail to reject H0 → No significant difference between A and B  '''
```

```python
# TEST 2 
## **H2 — Experiment-Level Lift** 
'''
**Hypothesis (per experiment)**

* **H₀ (Null):** Variant B does **not** improve conversion compared to Variant A for this experiment.
* **H₁ (Alternative):** Variant B has a **higher conversion rate** than Variant A for this experiment.

**Test Logic**

* Use a **two-proportion z-test** **per experiment**.
* Because we’re testing multiple experiments, apply a **Bonferroni correction** to adjust the p-values. 
  This prevents false positives (Type I errors).

**Metrics**

* Conversion rate per experiment: `num_converted / total_users`
* Count of conversions (`num_converted`) and total users (`conv_users`) for B vs A per experiment.

**Output**

* Z-statistic per experiment
* Corrected p-value per experiment
* Lift (%) per experiment

'''
```

```python
Exp_ConvSumm = analysisECOM_lvl.groupby(["experiment_id", "Experiment Version"]).agg(convperExp=("Total completed Purchases", "sum"), 
                                                                                        userperExp = ("user_id", "count")).reset_index()


Exp_ConvSumm["Conversion_Rate"] = Exp_ConvSumm["convperExp"]/Exp_ConvSumm["userperExp"]
```

```python
# USING Pivot tables we'll create the count and nobs 

exp_pivot = Exp_ConvSumm.pivot(

            index= "experiment_id",
            columns= "Experiment Version",
            values= ["convperExp","userperExp"]
)

exp_pivot

# SO NOW that we have all the values within one single table - we can use the convperExp and userperExp for A and B for each of the 8 experiments
```

```python
results = []

for exp in exp_pivot.index:

    A_conv = exp_pivot.loc[exp,("convperExp", "A")]
    B_conv = exp_pivot.loc[exp, ("convperExp", "B")]

    A_Users = exp_pivot.loc[exp,("userperExp", "A")]
    B_Users = exp_pivot.loc[exp,("userperExp", "B")]


    countconvT2 = [A_conv, B_conv]
    numofObsT2 = [A_Users, B_Users]


    z_stat, p_val = proportions_ztest(countconvT2, numofObsT2, alternative="two-sided")

    results.append({

        "experiment_id": exp,
        "Conv_Rate_A": A_conv/A_Users,
        "Conv_Rate_B": B_conv/B_Users,
        "Lift_B_vs_A": (B_conv/B_Users) - (A_conv/A_Users),
        "Z_Score": z_stat,
        "P_Value": p_val
    })


df_results = pd.DataFrame(results)

df_results
```

```python
Purchase_contingency_table = pd.crosstab(analysisECOM_lvl["Purchase_Intent_Level"], analysisECOM_lvl["Converted"])

Purchase_contingency_table
```

```python

'''
What this does:
chi2 → Chi-Square statistic (measures difference between observed vs expected frequencies)
p → p-value (if < 0.05, conversion depends on segment)
dof → degrees of freedom
expected → expected counts if no association existed 
'''
```

```python
from scipy.stats import chi2_contingency

# Compeleted Purchase Level
chi2, p, dof, expected = chi2_contingency(Purchase_contingency_table)

print(f"chi2:", chi2) 
print(f"p value:", p)
```

```python
'''
**Test 3 (Part D) — Purchase Intent Level × Conversion (Chi-Square)**

### **Result**
**Chi-square statistic:** 0.2941
**p-value:** 0.8633


### **Statistical Interpretation**
* The p-value (**0.863**) is **much greater than 0.05**
* We **fail to reject the null hypothesis**
* This means **conversion is NOT statistically dependent on Purchase Intent Level**

In simple terms:
Differences in conversion across **Low / Medium / High purchase intent users** are very likely due to **random variation**, not a real underlying relationship.


### **Business Interpretation**
* Purchase intent level **alone does not meaningfully predict conversion**
* High “intent” users are **not statistically more likely to convert** than low- or medium-intent users
* Targeting users *only* based on purchase intent would **not reliably increase conversions**

This is a **strong and honest negative result** — it shows:
* You tested a reasonable hypothesis
* You didn’t force significance
* You ruled out a commonly assumed driver

Exactly the kind of disciplined analysis recruiters respect.
'''
```

```python
import pandas as pd
import statsmodels.formula.api as smf
import numpy as np

# Ensure categorical variables have the correct reference levels
analysisECOM_lvl["Engagement_Level"] = pd.Categorical(
    analysisECOM_lvl["Engagement_Level"],
    categories=["Low", "Medium", "High"],
    ordered=True
)

analysisECOM_lvl["PagesView_Level"] = pd.Categorical(
    analysisECOM_lvl["PagesView_Level"],
    categories=["Low", "Medium", "High"],
    ordered=True
)

analysisECOM_lvl["ClickProd-Level"] = pd.Categorical(
    analysisECOM_lvl["ClickProd-Level"],
    categories=["Low", "Medium", "High"],
    ordered=True
)

analysisECOM_lvl["Purchase_Intent_Level"] = pd.Categorical(
    analysisECOM_lvl["Purchase_Intent_Level"],
    categories=["Low", "Medium", "High"],
    ordered=True
)

analysisECOM_lvl["Experiment Version"] = pd.Categorical(
    analysisECOM_lvl["Experiment Version"],
    categories=["A", "B"]
)
```

```python
model = smf.logit(
    formula="""
    Converted ~ 
    C(Engagement_Level) + 
    C(PagesView_Level) + 
    C(Q("ClickProd-Level")) + 
    C(Purchase_Intent_Level) +
    C(Q("Experiment Version"))
    """, data=analysisECOM_lvl
).fit()

print(model.summary())
```

```python
params = model.params
conf = model.conf_int()

odds_ratios = np.exp(params)
conf_odds = np.exp(conf)

results = pd.DataFrame({
    "Odds_Ratio": odds_ratios,
    "CI_Lower": conf_odds[0],
    "CI_Upper": conf_odds[1],
    "p_value": model.pvalues
})

results
```
